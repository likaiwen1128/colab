import libraries 

````
import pandas as pd
import altair as alt
import pandas_datareader.data as web 
from datetime import datetime, date 
````

apple data

````
start= datetime(2016,1,1)
end= datetime (2016,12,31)
````

````
apple=web.DataReader("AAPL",'yahoo',start,end)
apple.tail()
````

````
type(apple)
````

````
apple['month']=apple.index.month
apple.head()
````

````
apple['date1']=apple.index
apple.head()
````

Graph of price

````
alt.Chart(apple).mark_line(color='orange', strokeWidth=2).encode(x="date1:T",y="Close:Q")
````

````
alt.Chart(apple).mark_area(color='orange').encode(x="date1:T",y="Close:Q").interactive()
````

add average and percentage

````
apple['average']=(apple.High+apple.Low)/2
apple['percentage']=(apple.Close-apple.Open)/apple.Open
apple.head()
````

High vs. Low

````
c1=alt.Chart(apple).mark_line(color='blue',strokeWidth=2,strokeJoin='round').encode(x="date1:T",y="High:Q")
c2=alt.Chart(apple).mark_line(color='red',strokeWidth=2).encode(x="date1:T",y="Low:Q")
c1+c2
c1+c2.interactive()
````

percentage change

````
alt.Chart(apple).mark_square().encode(
    x='date1:T',
    y='percentage',
    color='Volume',
    size='Volume',
    tooltip=[alt.Tooltip('date1:T', format='%B %e')]
)
````

````
alt.Chart(apple).mark_bar(color='#e63293').encode(
    x=alt.X("percentage",bin=alt.Bin(maxbins=30)),
    y='count()',
    tooltip='count()'
)
````

````
apple['wd']=apple.index.weekday
apple.head()
````

percentage vs. weekday

````
alt.Chart(apple).mark_circle().encode(
    x='wd',
    y='percentage',
    size='Volume',
    color='Volume',
    tooltip=[alt.Tooltip('date1:T', format='%A %B %e'),alt.Tooltip('percentage')]
)
````

````
apple['hilow']=apple.High-apple.Low
apple.head()
````

daily difference vs. percentage change

````
alt.Chart(apple).mark_circle().encode(
    x='hilow',
    y='percentage',
    color='Volume',
    size='Volume',
    tooltip=[alt.Tooltip('date1:T', format='%B %e')]
)
````
